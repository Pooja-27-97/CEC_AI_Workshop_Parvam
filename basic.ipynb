```python
# basic data types and variables in Python

# integers
count = 10
age = 18

# floating point
price = 19.99
temperature = 36.6

# strings
name = "Johan"
message = 'Hello, Sukuna!'

# boolean
is_active = False
is_valid = False

# list
numbers = [1, 2, 3, 4, 5]

# tuple
coords = (10, 20)

# dictionary
person = {"name": "TonFah", "age": 25}

# set
unique_items = {"apple", "banana", "cherry"}

print(count, price, name, is_active, message)
print(numbers, coords, person, unique_items)
```

```python
# Conditional statements
if age > 18:
    print(f"{name} is an adult.")
elif age == 18:
    print(f"{name} is exactly 18.")
else:
    print(f"{name} is a minor.")

# Check boolean
if is_active:
    print("User is active.")
else:
    print("User is not active.")

# Loops
# For loop with list
for num in numbers:
    print(f"Number: {num}")

# For loop with dictionary
for key, value in person.items():
    print(f"{key}: {value}")

# While loop
i = 1
while i < count:
    print(f"Count: {i}")
    i += 1

# Loop with set
for item in unique_items:
    print(f"Unique item: {item}")
```
